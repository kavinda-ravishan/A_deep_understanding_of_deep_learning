# **Gradient descent in 1D**
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def fx(x):
    return 3*x**2 - 3*x + 4

def deriv(x):
    return 6*x - 3
```

```python
x = np.linspace(-2, 2, 2001)
training_epochs = 100
```

## G.D without changing the learning rate

```python
localMin = np.array([-2])

print(f'First estimate : {localMin}')

learning_rate = 0.01
modelParamsFixed = np.zeros((training_epochs, 3))

for i in range(training_epochs):
    grad = deriv(localMin)
    lr = learning_rate # non-adaptive learning rate
    localMin = localMin - learning_rate*grad
    modelParamsFixed[i, :] = localMin, grad, lr 

print(f'final : {localMin}')

plt.plot(x, fx(x))
plt.plot(modelParamsFixed[:,0], fx(modelParamsFixed[:,0]),  'bo', markersize=3)
plt.plot(0.5, fx(0.5), 'gs')

plt.xlim(x[[0, -1]])
plt.title('Empirical local minimum : %s'%localMin[0])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

plt.plot(modelParamsFixed[:,2], 'bo:', markersize=3)
plt.xlim([0, training_epochs])
plt.title('Learning Rate')
plt.xlabel('Epoch')
plt.ylabel('Learning rate')
plt.grid()
plt.show()
```

## G.D. using gradiaent-based learning rate

```python
localMin = np.array([-2])

print(f'First estimate : {localMin}')

learning_rate = 0.01
modelParamsGrad = np.zeros((training_epochs, 3))

for i in range(training_epochs):
    grad = deriv(localMin)
    lr = learning_rate*np.abs(grad) # adaptive learning rate according to the gradient
    localMin = localMin - learning_rate*grad
    modelParamsGrad[i, :] = localMin, grad, lr 

print(f'final : {localMin}')

plt.plot(x, fx(x))
plt.plot(modelParamsGrad[:,0], fx(modelParamsGrad[:,0]),  'bo', markersize=3)
plt.plot(0.5, fx(0.5), 'gs')

plt.xlim(x[[0, -1]])
plt.title('Empirical local minimum : %s'%localMin[0])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

plt.plot(modelParamsGrad[:,2], 'bo:', markersize=3)
plt.xlim([0, training_epochs])
plt.title('Learning Rate')
plt.xlabel('Epoch')
plt.ylabel('Learning rate')
plt.grid()
plt.show()
```

## G.D. using a time-based learning rate

```python
localMin = np.array([-2])

print(f'First estimate : {localMin}')

learning_rate = 0.1
modelParamsTime = np.zeros((training_epochs, 3))

for i in range(training_epochs):
    grad = deriv(localMin)
    lr = learning_rate * ( 1 - ((1+i)/training_epochs)) # adaptive learning rate according to the epoch
    localMin = localMin - learning_rate*grad
    modelParamsTime[i, :] = localMin, grad, lr 

print(f'final : {localMin}')

plt.plot(x, fx(x))
plt.plot(modelParamsTime[:,0], fx(modelParamsTime[:,0]),  'bo', markersize=3)
plt.plot(0.5, fx(0.5), 'gs')

plt.xlim(x[[0, -1]])
plt.title('Empirical local minimum : %s'%localMin[0])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

plt.plot(modelParamsTime[:,2], 'bo:', markersize=3)
plt.xlim([0, training_epochs])
plt.title('Learning Rate')
plt.xlabel('Epoch')
plt.ylabel('Learning rate')
plt.grid()
plt.show()
```
